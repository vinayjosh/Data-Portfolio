## Q22. What percentage of Miles were clocked under Business Category and what percentage of Miles were clocked under Personal Category ? (3 points)

### Note:Use the original dataframe without dropping the 'NA' values. 


```python
ub_per = ub[ub['CATEGORY*']=='Personal']
ub_per["MILES*"].sum()
```

```python
ub_bus = ub[ub['CATEGORY*']=='Business']
ub_bus["MILES*"].sum()
```

```python
ub["MILES*"].sum()
```

```python
miles_bus = (ub_bus["MILES*"].sum() / ub["MILES*"].sum())*100
miles_bus
```

```python
#94.11% of Miles covered fall under the Business Category of travel
```

```python
miles_per = (ub_per["MILES*"].sum() / ub["MILES*"].sum())*100
miles_per
```

```python
#5.8% of Miles covered fall under the Personal Category of travel
```
